```python
board = np.zeros((18, 11, 3), dtype=float)

print_dict = {
    0: "M",
    1: "S",
    2: "E"
}

def print_board_bar():
    for i in range(26):
        print("-", end="")
    print()

def print_board(board):
    print(" "*3, end="| ")
    for i in range(11):
        print(f"{chr(i + 65)}", end=" ")
    print()
    print_board_bar()
    for i in range(18):
        print(f"{18 - i:2} |", end=" ")
        for j in range(11):
            num_printed = False
            for c in range(2):
                if board[17 - i, j, c] > 0:
                    char = "E"
                    if c > 0:
                        char = "S" if i >= 12 else "E"
                    print(char, end=" ")  
                    num_printed = True
            if not num_printed:
                print("-", end=" ")
        print()
    print_board_bar()

def add_move(board, moveString, isStart, isEnd):
    row = int(moveString[1:]) - 1
    col = ord(moveString[0]) - 65
    if col >= 0 and col < 11: # matched starts are sometimes indicated with lowercase letters, but we don't need them (for now, may add them later)
        board[row, col, int(isStart) + 2 * int(isEnd)] = 1

for idx, move in movelist.iterrows():
    add_move(board, move.description, move.isStart, move.isEnd)

print_board(board)
```

```python
filtered = normalized[["grade", "userGrade", "userRating", "moves", "repeat_weight", "repeats"]]
filtered = filtered[filtered["repeats"] > 3]
filtered = filtered[filtered["userRating"] > 3]
filtered = filtered[filtered["grade"] != "8B+"]
```

```python
# One hot encoding method for use in ordinal regression networks
# Classification networks "Classify" based on the classes given, and are able to more directly learn the ordinality of the data
# Think of classes as P(Grade > x) for each x in list, (x0 = 6B+, x1 = 6C, and so on)
# There are so few examples of hard and repeated problems that they are combined together as 8A and up (seems reasonable from my experience on the moonboard anyways)
grade_dict = {
    "6B+": np.array([0,0,0,0,0,0,0]),
    "6C": np.array([1,0,0,0,0,0,0]), 
    "6C+": np.array([1,0,0,0,0,0,0]), # both 6C and 6C+ are V5
    "7A": np.array([1,1,0,0,0,0,0]),
    "7A+": np.array([1,1,1,0,0,0,0]),
    "7B": np.array([1,1,1,1,0,0,0]), # both 7B and 7B+ are V8
    "7B+": np.array([1,1,1,1,0,0,0]),
    "7C": np.array([1,1,1,1,1,0,0]),
    "7C+": np.array([1,1,1,1,1,1,0]),
    "8A": np.array([1,1,1,1,1,1,1]), # 8A and up are considered the same (very difficult and vary wildly in true difficulty)
    "8A+": np.array([1,1,1,1,1,1,1]),
    "8B": np.array([1,1,1,1,1,1,1]),
}

grade_num_dict = {
    "6B+": 0,
    "6C": 1, 
    "6C+": 1, # both 6C and 6C+ are V5
    "7A": 2,
    "7A+": 3,
    "7B": 4, # both 7B and 7B+ are V8
    "7B+": 4,
    "7C": 5,
    "7C+": 6,
    "8A": 7, # 8A and up are considered the same (very difficult and vary wildly in true difficulty, + not enough examples)
    "8A+": 7,
    "8B": 7,
}
```

```python
def board_from_moves(moves):
    moveList = pd.json_normalize(moves.iloc[0])[["description", "isStart", "isEnd"]]
    board_int = np.zeros((18, 11, 3), dtype=float)
    board_final = np.zeros((18, 11, 2), dtype=float)
    for idx, move in moveList.iterrows():
        add_move(board_int, move.description, move.isStart, move.isEnd)
    # adjust for two vs one hand start / finish
    board_int[:, :, 1] /= np.sum(board_int[:, :, 1])
    board_int[:, :, 2] /= np.sum(board_int[:, :, 2])
    board_final[:, :, 0] = board_int[:, :, 0]
    board_final[:, :, 1] = np.maximum(board_int[:, :, 1], board_int[:, :, 2])
    return board_final
```

```python
filtered["numberGrade"] = filtered["userGrade"].map(grade_dict)
filtered["numberGrade"].fillna(filtered["grade"].map(grade_dict), inplace=True)
filtered['board'] = filtered[["moves"]].apply(board_from_moves, axis=1)

filtered["grade_numeric"] = filtered["userGrade"].map(grade_num_dict)
filtered["grade_numeric"].fillna(filtered["grade"].map(grade_num_dict), inplace=True)
```
